```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

data = "LL-Output-Tue-12-Mar-2024-at-10-18-05PM.txt"
df = pd.read_csv(data, sep='\s+', skiprows=7)
df = df.drop(columns=['Energy:', 'Order:'])
df = df.rename(columns={'Ratio:' : 'Order', 'step:' : 'Energy', 'MC' : 'Ratio',  '#' : 'MC_step' })
df = df.drop(df.index[0])
df.iloc[1:, 0] = df.iloc[1:, 0].astype(str).str.lstrip('0')

df
```

```python
# Plotting change in energy over each MCS
plt.figure(figsize=(10, 5))
plt.plot(df['MC_step'], df['Energy'], linestyle='-')
plt.xlabel('MC_step')
plt.ylabel('Reduced Energy')
plt.title('MC Step vs Reduced Energy')

plt.xticks(df['MC_step'][::10])

plt.show()
```

```python
# Plotting MC_step v order Paramter S
plt.figure(figsize=(10, 5))
plt.plot(df['MC_step'], df['Order'], linestyle='-')
plt.xlabel('MC_step')
plt.ylabel('Order Parameter, S')
plt.title('MC Step v Order')


plt.xticks(df['MC_step'][::10])

plt.show()
```
